```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x **2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X_Label')
plt.ylabel('Y_Label')
plt.title('Title')
```

```python
plt.subplot(1,3,1)
plt.plot(x,y,'r')


plt.subplot(1,3,2)
plt.plot(x,y,'b')

plt.subplot(1,3,3)
plt.plot(y,x,'g')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes1 = fig.add_axes([0.2,0.5,0.4,0.3])

axes.plot(x,y,'b')
axes.set_xlabel('X_Label')
axes.set_ylabel('Y_Label')
axes.set_title('Title')

axes1.plot(y,x,'r')
axes1.set_xlabel('X_Label')
axes1.set_ylabel('Y_Label')
axes1
```
