```python
import cv2
import matplotlib.pyplot as plt
```

```python
cap = cv2.VideoCapture(0)
```

```python
if cap.isOpened():
    ret , frame = cap.read()
    print('ret'+str(ret))
    print(frame)
       
else:
    ret = False

img = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    
    
plt.imshow(img)
plt.title('Webcam Image Capture')
plt.xticks([])
plt.yticks([])
plt.show()


cap.release()    
```
